```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format="retina"  # only for retina display
```

```python
task_1 = open("task1", 'r')
# task_3 = open("task3", 'r')
# task_4 = open("task4", 'r')
# task_5 = open("task5", 'r')
# task_6 = open("task6", 'r')
# task_7 = open("task7", 'r')
```

```python
lines = task_1.readlines()
task_1.close()
lines = np.array(lines)
```

```python
def BuildGraph(data, oper_name):
    plt.figure(figsize=(20,12))
    block_size = []
    array_size = data[0].split()[0]
    times = []
    
    for d in data:
        splitted = [float(x) for x in d.split()]
        times.append(splitted[-1])
        block_size.append(splitted[1])

    plt.plot(block_size, times, linewidth=4)
    
    plt.title("Зависимость времени работы "+np.str(oper_name)+" размера "+np.str(array_size) + " от размера блока", fontsize=25)
    plt.xlabel("размер блока", fontsize=20)
    plt.ylabel("время, мс", fontsize=20)
    plt.show()
    
    
prev_index = 0    
for i in range(6, lines.size+1, 6):
    BuildGraph(lines[prev_index:i],"сложения векторов")
    prev_index = i
```

```python
task_2 = open("task2", 'r')
lines = task_2.readlines()
task_2.close()
lines = np.array(lines)

prev_index = 0    
for i in range(6, lines.size+1, 6):
    BuildGraph(lines[prev_index:i],"поэлементного умножения векторов")
    prev_index = i
```
